# **Metodología empleada de scraping: Beautifulsoup**

```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Instalación de beautifulsoup4
!pip install beautifulsoup4

def scrape_mercadolibre(url):
    """
    Extrae información de los productos más vendidos de una tableta-samsung-galaxy-tab-s6 en Mercado Libre.

    Args:
        url: La URL de la página de Mercado Libre a raspar.

    Returns:
        Un string con la información extraída o un mensaje de error.
    """
    try:
        # 1. Obtener el contenido HTML de la página mercado libre
        response = requests.get(url)
        response.raise_for_status()  # excepción para códigos de estado incorrectos (4xx o 5xx)

        # Analizar el HTML con BeautifulSoup
        soup = BeautifulSoup(response.content, "html.parser")

        productos = soup.find_all("li", class_="ui-search-layout__item shops__layout-item")

        resultados = ""
        tableta = ""  # Inicializar la variable para evitar errores
        titulos = []
        precios = []
        valoraciones = []
        envio_gratis = []
        tableta_usada = []

        for producto in productos:
            # Título del producto
            titulo_elemento = producto.find("h2", class_="poly-box poly-component__title")
            titulo = titulo_elemento.text.strip() if titulo_elemento else "Título no encontrado"

            # Precio del producto
            precio_elemento = producto.find("span", class_="andes-money-amount__fraction")
            precio = precio_elemento.text.strip() if precio_elemento else "Precio no encontrado"

            # Valoración del producto
            valoracion_elemento = producto.find("span", class_="poly-reviews__rating")
            valoracion = valoracion_elemento.text.strip() if valoracion_elemento else "Valoración no encontrada"

            # Envío gratis
            envio_gratis_elemento = producto.find("div", class_="poly-component__shipping")
            envio = envio_gratis_elemento.text.strip() if envio_gratis_elemento else "Envío no encontrado"

            # Tableta usada
            tableta_usada_elemento = producto.find("span", class_="poly-component__item-condition")
            usada = tableta_usada_elemento.text.strip() if tableta_usada_elemento else "Nuevo"

            # Información a la cadena de resultados
            resultados += f"Título: {titulo}\n"
            resultados += f"Precio: {precio}\n"
            resultados += f"Valoración: {valoracion}\n"
            resultados += f"Envío: {envio}\n"
            tableta += f"Tableta: {usada}\n"
            resultados += "-----\n"

            # Guardar valores
            titulos.append(titulo)
            precios.append(precio)
            valoraciones.append(valoracion)
            envio_gratis.append(envio)
            tableta_usada.append(usada)

        # Guardar el HTML en el directorio especificado
        if not os.path.exists('content/sample_data'):
            os.makedirs('content/sample_data')  # Crea el directorio si no existe
        with open("content/sample_data/mercadolibre_data.html", "w", encoding="utf-8") as f:
            f.write(str(soup))  # Guarda el HTML completo

        # Crear dataframe
        df = pd.DataFrame({
            "Título": titulos,
            "Precio": precios,
            "Valoración": valoraciones,
            "Envío": envio_gratis,
            "Tableta": tableta_usada,
        })

        # Crear CSV
        df.to_csv('mercadolibre_producto_resultados.csv', index=False, encoding='utf-8')

        return resultados

    except requests.exceptions.RequestException as e:
        return f"Error al obtener la URL: {e}"
    except AttributeError as e:
        return f"Error al analizar el HTML: {e}. Verifica el selector CSS."
    except Exception as e:
        return f"Error inesperado: {e}"

# URL Samsung Galaxy modelo Tab S6 en Mercado Libre Colombia
url_mercadolibre = "https://listado.mercadolibre.com.co/tableta-samsung-galaxy-tab-s6"

resultados = scrape_mercadolibre(url_mercadolibre)
print(resultados)

```

# **df , csv**

```python
df = pd.read_csv('mercadolibre_producto_resultados.csv')

print(df.head())
```

# **PRODUCTOS MAS CAROS**

```python
productos_mas_caros = df.sort_values(by='Precio', ascending=False).head(5)
print(productos_mas_caros)
```

# **PROMEDIO DE LOS PRECIOS**

```python
df['Precio'] = pd.to_numeric(df['Precio'], errors='coerce')

# precio promedio
precio_promedio = df['Precio'].mean()
print(f"El precio promedio es: {precio_promedio:.2f}")
```

# **CONDICION NUEVO O USADO**

```python
condicion_count = df['Tableta'].value_counts()
print(condicion_count)
```

# **PRECIO PROMEDIO VALORACION**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# valoración para calcular precio promedio
precio_valoracion = df.groupby('Valoración')['Precio'].mean().reset_index()

# gráfico
sns.barplot(data=precio_valoracion, x='Valoración', y='Precio')
plt.title('Precio Promedio por Valoración')
plt.xlabel('Valoración')
plt.ylabel('Precio Promedio')
plt.show()
```

# **DISTRIBUCION PRECIO**

```python
# Histograma de precios
plt.figure(figsize=(8, 6))
sns.histplot(df['Precio'], bins=20, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

# **NUEVOS Y USADOS**

```python
condicion_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6))
plt.title('Porcentaje de Productos por Condición')
plt.ylabel('')
plt.show()
```

# **PRECIO Y LA VALORIZACION**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Valoración', y='Precio', hue='Tableta')
plt.title('Relación entre Precio y Valoración')
plt.xlabel('Valoración')
plt.ylabel('Precio')
plt.legend(title='Condición')
plt.show()
```

# **GRAFICO CONBINADO**

```python
# productos con envío gratis
df['Envío Gratis'] = df['Envío'].str.contains("gratis", case=False, na=False)

# envío con el  calcular precios promedio
envio_precio = df.groupby('Envío Gratis')['Precio'].mean().reset_index()

# gráfico
sns.barplot(data=envio_precio, x='Envío Gratis', y='Precio', palette='viridis')
plt.title('Precio Promedio según Envío Gratis')
plt.xlabel('Envío Gratis')
plt.ylabel('Precio Promedio')
plt.xticks([0, 1], ['No', 'Sí'])
plt.show()
```
